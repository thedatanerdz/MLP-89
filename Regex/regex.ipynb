<h1 align='center'>NLP concept: Regular Expressions</h1>

**What is it**
Regular expressions (regex) are patterns used to search, match, and manipulate text data in machine learning. They are helpful for tasks like finding specific words, validating email addresses, extracting data from text, or cleaning and preprocessing text before feeding it into models.

**How does it work**
Regex works by defining a pattern composed of characters, symbols, and special sequences that describe the desired text structure. When applied to a given input string, regex engines search for occurrences of the defined pattern and return matching portions of the text. It's like a powerful find-and-replace tool that allows you to locate and manipulate text based on complex rules, enabling tasks such as data extraction, validation, and transformation.

**How is it used in industry**
- Data Validation and Cleaning: Regex is used to validate and clean data, such as ensuring proper formatting of phone numbers, email addresses, credit card numbers, and other structured information.
- Text Extraction: It's employed to extract specific information from unstructured text data, like extracting dates, addresses, product names, or relevant keywords from documents.
- Search and Replace: Regex is used for searching and replacing specific patterns in large amounts of text, enabling efficient text editing and manipulation.
- Web Scraping: In web development and data mining, regex helps extract data from HTML content by targeting specific tags, classes, or patterns.
- Log Analysis: Regex is used to parse and analyze log files to identify errors, anomalies, or specific events, aiding in troubleshooting and performance optimization.
- Natural Language Processing (NLP): Regex can be used in basic text preprocessing for tokenization, removing punctuation, or identifying patterns like hashtags, mentions, or URLs.
- Data Transformation: In ETL (Extract, Transform, Load) pipelines, regex is employed to transform raw data into structured formats before loading into databases or other systems.
- Code Refactoring: Developers use regex to search and replace patterns in code, making it easier to refactor and update codebases.
- Security and Compliance: Regex is used in cybersecurity and data privacy to identify sensitive information like social security numbers or credit card details in unstructured text, aiding in compliance with regulations.
- Automated Testing: In software testing, regex can help verify whether output strings match expected patterns, enhancing test coverage.
- Information Retrieval: In search engines and information retrieval systems, regex can be used to filter and rank search results based on relevance to user queries.

<h3>(1) Regex in customer support</h3>

<h3>Retrieve order number</h3>

```python
import re

chat1='codebasics: Hello, I am having an issue with my order # 412889912'

pattern = 'order[^\d]*(\d*)'
matches = re.findall(pattern, chat1)
matches
```

```python
chat2='codebasics: I have a problem with my order number 412889912'
pattern = 'order[^\d]*(\d*)'
matches = re.findall(pattern, chat2)
matches
```

```python
chat3='codebasics: My order 412889912 is having an issue, I was charged 300$ when online it says 280$'
pattern = 'order[^\d]*(\d*)'
matches = re.findall(pattern, chat3)
matches
```

```python
def get_pattern_match(pattern, text):
    matches = re.findall(pattern, text)
    if matches:
        return matches[0]
```

```python
get_pattern_match('order[^\d]*(\d*)', chat1)
```

```python
def extract_personal_information(text):
    age = get_pattern_match('age (\d+)', text)
    full_name = get_pattern_match('Born(.*)\n', text)
    birth_date = get_pattern_match('Born.*\n(.*)\(age', text)
    birth_place = get_pattern_match('\(age.*\n(.*)', text)
    return {
        'age': int(age),
        'name': full_name.strip(),
        'birth_date': birth_date.strip(),
        'birth_place': birth_place.strip()
    }
```
